```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# 2D Ising schematic phase transition plot
# Magnetization vs temperature
# Units: J = 1, k_B = 1
# -----------------------------

J = 1.0
kB = 1.0

# Critical temperature for 2D square-lattice Ising model
Tc = 2 * J / (kB * np.log(1 + np.sqrt(2)))

# Temperature range
T = np.linspace(0.2, 5.0, 800)

# Exact spontaneous magnetization for 2D Ising model at zero field
M = np.zeros_like(T)
below_Tc = T < Tc
M[below_Tc] = (1 - np.sinh(2 * J / (kB * T[below_Tc]))**(-4))**(1/8)

# -----------------------------
# Plot
# -----------------------------

fig, ax = plt.subplots(figsize=(9, 5.5), dpi=300)

# Shaded phase regions
ax.axvspan(T.min(), Tc, color="#4C78A8", alpha=0.14)
ax.axvspan(Tc, T.max(), color="#F58518", alpha=0.10)

# Magnetization curve
ax.plot(T, M, color="black", linewidth=3.5)

# Critical temperature line
ax.axvline(Tc, color="black", linestyle="--", linewidth=2.3, alpha=0.8)

# Critical point marker
ax.scatter([Tc], [0], s=130, color="black", zorder=5)

# Region labels
ax.text(
    1.05, 0.73,
    "Ordered phase",
    fontsize=18,
    
    ha="center",
    va="center"
)

ax.text(
    3.75, 0.38,
    "Disordered phase",
    fontsize=18,
    
    ha="center",
    va="center"
)

# Critical point annotation
ax.annotate(
    "Critical point ($T_c$)",
    xy=(Tc, 0),
    xytext=(Tc + 0.35, 0.18),
    fontsize=17,
    ha="left",
    va="center",
    arrowprops=dict(
        arrowstyle="->",
        linewidth=2,
        color="black"
    )
)

# Optional analogy labels for your privacy talk
ax.text(
    1.05, 0.18,
    "structure recoverable",
    fontsize=16,
    ha="center",
    va="center"
)

ax.text(
    3.75, 0.10,
    "structure hidden",
    fontsize=16,
    ha="center",
    va="center"
)

# Axis labels and title
ax.set_title(
    "Order–Disorder Transition: Analogy for Privacy Failure",
    fontsize=20,
    
    pad=10
)

ax.set_xlabel("Temperature (T)", fontsize=18, labelpad=10)
ax.set_ylabel("Magnetization (M)", fontsize=18, labelpad=10)

# Limits
ax.set_xlim(T.min(), T.max())
ax.set_ylim(-0.03, 1.05)

# Ticks
ax.set_xticks([1, 2, Tc, 3, 4, 5])
ax.set_xticklabels(["1", "2", r"$T_c$", "3", "4", "5"], fontsize=16)

ax.tick_params(axis="y", labelsize=16)
ax.tick_params(axis="both", width=1.6, length=6)

# Grid and spines
ax.grid(True, alpha=0.25, linewidth=1)

ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_linewidth(1.6)
ax.spines["bottom"].set_linewidth(1.6)

plt.tight_layout()

# Save for slide use
plt.savefig("ising_order_disorder_privacy_analogy.png", dpi=300, bbox_inches="tight")

plt.show()
```
